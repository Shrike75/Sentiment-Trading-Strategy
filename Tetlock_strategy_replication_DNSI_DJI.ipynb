# Replication: Tetlock (2007) — Media Pessimism and Stock Market Returns  


## 1. Paper Summary


In “Giving Content to Investor Sentiment: The Role of Media in the Stock Market,” Paul C. Tetlock (published in The Journal of Finance) examines whether the tone of financial media can help explain short-horizon stock-market behavior by extracting a “pessimism” signal from the Wall Street Journal’s Abreast of the Market column. He argues that media pessimism contains measurable information about investor sentiment that can move prices and trading activity in the short run, beyond what is captured by standard market variables. Using automated content analysis and time-series regressions, he shows that unusually pessimistic language is associated with an immediate market decline that later reverses, and that extreme pessimism is associated with elevated trading volume. He also documents that market moves themselves predict next-day media tone, indicating feedback between returns and media instead of just one-way causality. The overall takeaway is that media sentiment is a quantifiable channel through which investor psychology and liquidity forces can generate temporary price movements and abnormal trading.

The paper’s first key contribution is methodological: Tetlock presents a unique way to convert daily financial commentary into a numerical sentiment series suitable for econometric testing. He applies the General Inquirer dictionary to each day’s “Abreast of the Market” column, counting category-specific word frequencies and then adjusting those frequencies to remove systematic day-of-week patterns. To reduce the many overlapping dictionary categories into a single interpretable measure, he uses principal components factor analysis and treats the first factor as a “pessimism” index. This is effectively a weighted combination of language categories that captures the dominant shared movement in negative tone. This is important because it avoids relying on an arbitrary word list for a single hypothesis and instead produces a data-driven sentiment proxy. The takeaway from is that “sentiment” becomes an observable time series built from transparent text features, making it possible to test precise predictions about returns and volume.

The main empirical claim is that pessimistic media tone forecasts short-run returns in a way consistent with temporary mispricing rather than lasting fundamental news. In predictive regressions (and VAR-style specifications), higher pessimism is followed by lower next-day market returns, but those losses are mostly reversed over the following days (roughly within a week), implying that the initial effect is temporary. Tetlock emphasizes that this pattern is difficult to reconcile with a pure “fundamentals information” story, because genuine negative information about cash flows shouldn’t systematically reverse, and it is also inconsistent with the notion that the column merely echoes already-priced information, because it retains predictive power for short-term returns. He also shows that simpler negative-language proxies behave similarly and argues that the magnitude is too large to be dismissed as a microstructure artifact. The takeaway is that the return response’s distinctive “impact then reversal” pattern is the paper’s main evidence for sentiment-driven price pressure.

A second piece of the evidence is the link between sentiment and trading volume, which Tetlock uses to discriminate among competing theories. He argues that if pessimism is a reflection of disagreement or liquidity shocks from noise trading, then extreme tone (captured by the absolute value of pessimism) should predict higher trading volume because more investors are incentivized to rebalance or because liquidity demand rises. In volume regressions that include lags of returns, pessimism, and |pessimism|, he finds that unusually large absolute pessimism robustly forecasts increases in next-day volume across multiple sentiment measures. This result strengthens the behavioral/liquidity interpretation because an “information proxy” view does not naturally imply that sentiment extremes should systematically raise volume, and a “media is irrelevant” view would predict no relationship at all. The takeaway is that volume is a core piece of the argument that media tone tracks (and possibly amplifies) shifts in sentiment and liquidity demand.

Finally, the paper treats the media–market relationship as a dynamic system with feedback in both directions and uses that fact to sharpen interpretation. Because the column is written around the market close data and often feels like a narrative explanation of the day, Tetlock shows that negative market returns predict more pessimistic tone in the next day’s coverage, meaning the sentiment measure is partly reactive even while it forecasts short-run market dynamics. He then argues that the combined evidence of short-run return pressure with reversal, stronger volume on sentiment extremes, and return-to-media feedback, fits best with noise and liquidity trader models, and is less consistent with fundamentals-information or volatility-proxy explanations. The takeaway is that quantifying media tone provides a disciplined way to study how investor psychology can temporarily move markets, and it offers a clear blueprint for replication and modern extensions using newer text representations. I’m quite interested in how individual investor behavior can be modeled and its effects on market movements, so this was a fascinating paper to read and replicate.

## 2. Replication targets (hypotheses and tests)
I replicate Tetlock’s central empirical predictions in a daily-lag framework with **five** lags (one trading week), using **HAC (Newey–West)** inference and standard market controls.

### H1 — Return pressure + reversal (sentiment interpretation)
**Hypothesis:** Higher pessimism at day *t−1* predicts **lower** next-day market returns, followed by a **reversal** over days *t+2* to *t+5*, so that the **sum** of the five pessimism coefficients is close to zero.  
**Tests:**  
1) Joint significance of pessimism lags in the return equation (Wald/F-test).  
2) Sign pattern: negative coefficient on lag 1 and positive cumulative effect on lags 2–5.  
3) “Reversal” test: whether the sum of lags 2–5 offsets lag 1; and whether the sum of lags 1–5 differs from zero.

### H2 — Volume responds to extremes
**Hypothesis:** Unusually high or low pessimism predicts **higher** next-day trading volume.  
**Test:** Regress volume (or a volume proxy) on lags of **absolute** pessimism, controlling for lagged volume, returns, and calendar effects.

### H3 — Stronger / more persistent effects for small stocks
**Hypothesis:** Negative sentiment predicts **more persistent** negative returns for small stocks than for large stocks.  
**Test:** Predict daily SMB (Fama–French) returns using pessimism lags and standard controls; examine whether the cumulative effect across a week remains negative.

### Extensions
1) Repeat H1/H2/H3 on **more recent data** (post-2000).  
2) Repeat H1 on **other asset classes** (bonds, gold, crypto) to test whether the effect is equity-specific or broader.  


# Literature Review

**De Long, J. Bradford; Shleifer, Andrei; Summers, Lawrence H.; Waldmann, Robert J. (1990a).** *Noise Trader Risk in Financial Markets.* **Journal of Political Economy.**  
**Thesis:** Non-fundamental (“noise”) beliefs can move prices away from fundamentals because rational arbitrage is risky and limited.  
**Main points:** The model formalizes how noise traders can create persistent mispricing and an additional risk (“noise trader risk”) that deters arbitrage. Sentiment shocks can generate temporary price pressure and predictable return patterns (including reversals). The framework helps explain why “belief shocks” can affect prices even when fundamentals are unchanged.  
**Relevance:** Provides the core theoretical mechanism behind Tetlock-style interpretations of media pessimism as sentiment-driven price pressure with later correction.


**Campbell, John Y.; Grossman, Sanford J.; Wang, Jiang (1993).** *Trading Volume and Serial Correlation in Stock Returns.* **Quarterly Journal of Economics**, 108, 905–939.  
**Thesis:** Liquidity and discount-factor shocks can generate predictable return components and link trading volume to return autocorrelation.  
**Main points:** The paper connects variation in liquidity demand/risk-bearing capacity to both volume and return predictability. It explains how temporary price pressure can arise alongside unusual trading volume. It provides an equilibrium lens for interpreting short-horizon serial correlation patterns.  
**Relevance:** Supports Tetlock’s empirical focus on joint dynamics of tone, returns, and volume, and motivates interpreting volume spikes as liquidity/participation shifts.

**Baker, Malcolm; Wurgler, Jeffrey (2006).** *Investor Sentiment and the Cross-Section of Stock Returns.* **The Journal of Finance**, 61(4), 1645–1680.  
**Thesis:** Investor sentiment drives return predictability, especially for stocks that are hard to value and hard to arbitrage.  
**Main points:** The authors construct a broad sentiment index and show that sentiment waves predict cross-sectional patterns: the effect is strongest in speculative or difficult-to-arbitrage stocks (small, young, distressed, etc.). The evidence aligns with limits-to-arbitrage mechanisms that allow mispricing to persist. The paper helped standardize “sentiment” as a measurable state variable in asset pricing.  
**Relevance:** Tetlock’s market-wide text-based pessimism measure lives in the same “sentiment state variable” family, and Baker–Wurgler motivates heterogeneity/limits-to-arbitrage robustness checks. Also motivates how sentiment can have larger affects on singular stock movements.


**Cutler, David M.; Poterba, James M.; Summers, Lawrence H. (1989).** *What Moves Stock Prices?* **Journal of Portfolio Management**, 15, 4–12.  
**Thesis:** A large share of short-horizon stock price movements cannot be cleanly linked to identifiable fundamental news.  
**Main points:** The paper is often cited as evidence that “news” explains less daily price variation than intuition suggests. This motivates behavioral, microstructure, attention, and sentiment channels alongside fundamentals. It also motivates better measurement of information and narratives in markets.  
**Relevance:** Provides motivation for why a media-tone measure might predict returns/volume even when fundamentals are hard to pin down.

**Engelberg, Joseph; Parsons, Christopher A. (2011).** *The Causal Impact of Media in Financial Markets.* **The Journal of Finance**, 66(1), 67–97.  
**Thesis:** Media coverage has a causal effect on trading behavior, not merely a passive reflection of information.  
**Main points:** Using plausibly exogenous variation in local news coverage of earnings announcements, the authors show that coverage predicts local trading after controlling for fundamentals. The design aims to separate attention/coverage effects from information content. Results support “media as attention catalyst” mechanisms.  
**Relevance:** Tetlock (2007) is primarily predictive/correlational; Engelberg–Parsons is a key complement for discussing identification and whether media induces (vs. reflects) market behavior.

**Antweiler, Werner; Frank, Murray Z. (2004).** *Is All That Talk Just Noise? The Information Content of Internet Stock Message Boards.* **The Journal of Finance**, 59, 1259–1294.  
**Thesis:** Online discussion contains both noise and information; its content relates to returns and volatility/volume proxies.  
**Main points:** The paper operationalizes text from message boards and typically finds small predictability for returns but stronger links to volume and volatility. It highlights practical measurement issues: spam/noise, selection bias, and separating sentiment from information. It broadens “text in markets” beyond professional media.  
**Relevance:** Methodologically similar (text → quantitative measure → market outcomes) and a useful robustness inspiration for alternative corpora and measurement pitfalls.

## Replication implications

- **Mechanism consistency:** Noise-trader and liquidity/attention frameworks predict temporary price pressure and potential reversal; replicate whether pessimism predicts short-horizon return pressure and subsequent reversion, and whether it predicts abnormal volume.
- **Media = information vs. attention:** Use identification logic from causal media studies (e.g., Engelberg–Parsons) and attention proxies (e.g., search intensity) to discuss whether results reflect information revelation, attention shocks, or sentiment waves.
- **Dictionary sensitivity (possible refutation):** Finance-specific dictionary critiques imply GI-based negativity may be mismeasured; could include robustness checks with alternative dictionaries and report sensitivity of coefficients/significance.

## 4. Load DNSI (FRBSF) and construct pessimism series
The DNSI file is distributed as an Excel spreadsheet by FRBSF. The code below downloads the file (cached locally), loads the daily series, and constructs a standardized pessimism variable.


```python
import requests

DNSI_URL = "https://www.frbsf.org/wp-content/uploads/news_sentiment_data.xlsx"
DNSI_XLSX_PATH = DATA_DIR / "news_sentiment_data.xlsx"

def download_if_missing(url: str, path: Path):
    if path.exists():
        return
    r = requests.get(url, timeout=60)
    r.raise_for_status()
    path.write_bytes(r.content)

download_if_missing(DNSI_URL, DNSI_XLSX_PATH)

def load_dnsi(path: Path) -> pd.DataFrame:
    xls = pd.ExcelFile(path)
    sheet = None
    for s in xls.sheet_names:
        if "data" in s.lower() or "index" in s.lower() or "dnsi" in s.lower():
            sheet = s
            break
    if sheet is None:
        sheet = xls.sheet_names[0]
    df = pd.read_excel(path, sheet_name=sheet)
    cols = {c.lower(): c for c in df.columns}
    date_col = next((cols[k] for k in cols if "date" in k), None)
    val_col = next((cols[k] for k in cols if "sentiment" in k), None)
    if date_col is None or val_col is None:
        print(date_col)
        print(val_col)
        raise ValueError("Unexpected column names")
    out = df[[date_col, val_col]].rename(columns={date_col: "date", val_col: "dnsi"})
    out["date"] = pd.to_datetime(out["date"])
    out = out.dropna().sort_values("date").drop_duplicates("date")
    return out

dnsi = load_dnsi(DNSI_XLSX_PATH)
dnsi.head(), dnsi.tail()
```

```python
# Construct pessimism (higher = more negative tone) and standardize
dnsi = dnsi.set_index("date")
dnsi["pessimism_raw"] = -dnsi["dnsi"].astype(float)
dnsi["pessimism"] = (dnsi["pessimism_raw"] - dnsi["pessimism_raw"].mean()) / dnsi["pessimism_raw"].std()
dnsi[["dnsi","pessimism_raw","pessimism"]].describe()
```

## 5. Load market data (Yahoo Finance) and compute returns
I download adjusted daily prices and (where relevant) volumes. Returns are computed as daily percentage returns.


```python
def download_prices(tickers, start, end=None):
    px = yf.download(tickers, start=start, end=end, auto_adjust=True, progress=False)
    if isinstance(px.columns, pd.MultiIndex):
        adj = px["Close"].copy()
        vol = px["Volume"].copy() if "Volume" in px else None
    else:
        adj = px["Close"].to_frame(name=tickers[0])
        vol = px["Volume"].to_frame(name=tickers[0]) if "Volume" in px else None
    adj.index = pd.to_datetime(adj.index)
    if vol is not None:
        vol.index = pd.to_datetime(vol.index)
    return adj, vol

# Core replication assets
core_tickers = ["^DJI", "DIA", "IWM", "^GSPC", "QQQ", "TLT", "GLD", "BTC-USD"]
prices, volumes = download_prices(core_tickers, start="1984-01-01", end=EXT_END)
prices.head(), prices.tail()
```

```python
rets = prices.pct_change().rename(columns=lambda c: f"ret_{c}")
rets = rets.replace([np.inf, -np.inf], np.nan)

# Volume proxies
vol = None
if volumes is not None:
    vol = volumes.rename(columns=lambda c: f"vol_{c}").replace([np.inf, -np.inf], np.nan)
rets.describe().T.head()
```

## 6. Merge sentiment with returns and construct controls
I align DNSI with market days and construct:
- calendar dummies (day-of-week, January),
- lagged returns and lagged volume,
- a volatility proxy via detrended squared residuals (paper-inspired).


```python
panel = dnsi[["pessimism"]].join(rets, how="inner")
if vol is not None:
    panel = panel.join(vol, how="left")

panel = panel.sort_index()
panel.head(), panel.shape
```

```python
def add_calendar_controls(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy()
    dow = out.index.dayofweek
    for k, name in enumerate(["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]):
        out[f"dow_{name}"] = (dow == k).astype(int)
    # January dummy
    out["is_jan"] = (out.index.month == 1).astype(int)
    # Keep only Mon-Fri in dummies, and drop Monday as baseline
    keep = ["dow_Tue","dow_Wed","dow_Thu","dow_Fri","is_jan"]
    return out, keep

panel, CAL_CONTROLS = add_calendar_controls(panel)
CAL_CONTROLS
```

```python
def lag_matrix(series: pd.Series, lags: int = 5, prefix: str = "") -> pd.DataFrame:
    return pd.concat({f"{prefix}L{k}": series.shift(k) for k in range(1, lags+1)}, axis=1)

def prepare_regression_frame(df: pd.DataFrame, y_col: str, x_cols: list, lags: int = 5) -> pd.DataFrame:
    d = df.copy()
    # Build lagged X blocks columns passed in x_cols
    X_parts = []
    for col in x_cols:
        X_parts.append(lag_matrix(d[col], lags=lags, prefix=f"{col}_"))
    X = pd.concat(X_parts, axis=1)
    X = pd.concat([X, d[CAL_CONTROLS]], axis=1)
    X = sm.add_constant(X, has_constant="add")
    out = pd.concat([d[[y_col]], X], axis=1)
    return out
```

## 7. Core return predictability regressions (5 lags)
I estimate OLS regressions with Newey–West (HAC) standard errors (lag=5). This is equivalent to the Granger-causality interpretation used in the paper when estimating each equation separately.


```python
def ols_hac(y: pd.Series, X: pd.DataFrame, hac_lags: int = 5):
    model = sm.OLS(y, X, missing="drop")
    res = model.fit()
    cov = cov_hac(res, nlags=hac_lags)
    res_hac = res.get_robustcov_results(cov_type="HAC", maxlags=hac_lags)
    return res, res_hac

def wald_joint_test(res_hac, param_names: list):
    # H0: all listed params = 0
    R = np.zeros((len(param_names), len(res_hac.params)))
    pname_to_idx = {n:i for i,n in enumerate(res_hac.model.exog_names)}
    for j,p in enumerate(param_names):
        R[j, pname_to_idx[p]] = 1.0
    q = np.zeros(len(param_names))
    return res_hac.wald_test((R, q))

def coeff_sum_test(res_hac, param_names: list):
    # H0: sum(params) = 0
    R = np.zeros((1, len(res_hac.params)))
    pname_to_idx = {n:i for i,n in enumerate(res_hac.model.exog_names)}
    for p in param_names:
        R[0, pname_to_idx[p]] = 1.0
    q = np.zeros(1)
    return res_hac.wald_test((R, q))

def basis_points(x):
    return 10000 * x
```

### 7.1 H1 on the paper-period window (1984–1999)
Dependent variable: **Dow return** (`ret_^DJI`).  
Key regressors: 5 lags of standardized pessimism.  
Controls: 5 lags of Dow returns, and 5 lags of `vol_DIA` (detrended log volume proxy).


```python
def detrended_log_volume(vol_series: pd.Series) -> pd.Series:
    # Remove rolling mean to reduce slow-moving growth in volume
    v = np.log(vol_series.replace(0, np.nan))
    return v - v.rolling(252, min_periods=50).mean()

# Detrended log-volume proxy
vol_cols = [c for c in ["vol_^DJI", "vol_^GSPC", "vol_DIA"] if c in panel.columns]
if len(vol_cols) > 0:
    vol_base = panel[vol_cols].bfill(axis=1).iloc[:, 0]
    panel["vlm"] = detrended_log_volume(vol_base)
else:
    panel["vlm"] = np.nan

# Add volatility proxy: squared residuals from AR(5) on returns (from paper)
def volatility_proxy(ret: pd.Series) -> pd.Series:
    d = pd.DataFrame({"r": ret})
    X = sm.add_constant(lag_matrix(d["r"], 5, prefix="r_"), has_constant="add")
    res = sm.OLS(d["r"], X, missing="drop").fit()
    eps2 = res.resid**2
    return eps2 - eps2.rolling(252, min_periods=50).mean()

panel["vol_proxy"] = volatility_proxy(panel["ret_^DJI"])
paper = panel.loc[(panel.index >= PAPER_START) & (panel.index <= PAPER_END)].copy()
paper.shape
```

```python
# Build regression frame
lags = 5
y_col = "ret_^DJI"
x_base = ["pessimism", "ret_^DJI", "vlm", "vol_proxy"]
reg_df = prepare_regression_frame(paper, y_col=y_col, x_cols=x_base, lags=lags).dropna()

y = reg_df[y_col]
X = reg_df.drop(columns=[y_col])

res, res_hac = ols_hac(y, X, hac_lags=lags)
print(res_hac.summary())
```

```python
# H1 tests
pess_lags = [f"pessimism_L{k}" for k in range(1, lags+1)]
test_joint = wald_joint_test(res_hac, pess_lags)
test_sum_all = coeff_sum_test(res_hac, pess_lags)
test_sum_2_5 = coeff_sum_test(res_hac, [f"pessimism_L{k}" for k in range(2, lags+1)])

print("Joint test (all 5 pessimism lags = 0):", test_joint)
print("Sum test (lags 1-5 sum = 0):", test_sum_all)
print("Reversal component (lags 2-5 sum = 0):", test_sum_2_5)

params = pd.Series(res_hac.params, index=res_hac.model.exog_names)
bse = pd.Series(res_hac.bse,    index=res_hac.model.exog_names)
coefs = params.loc[pess_lags]
ses = bse.loc[pess_lags]

print("\nPessimism lag coefficients (bp per 1 sd pessimism):")
for k in range(1, lags+1):
    name = f"pessimism_L{k}"
    beta_bp = 10000.0 * coefs[name]
    se_bp = 10000.0 * ses[name]
    print(f"  L{k}: {beta_bp:+.2f} bp (SE {se_bp:.2f})")
```

### Interpretation (H1)

In this H1 regression, lagged DNSI pessimism does not significantly predict next-day DJI returns once we include Tetlock-style controls and HAC(5) inference (joint test p = 0.904; no individual lag significant), but the pattern of point estimates is still directionally compatible with Tetlock’s story: the largest effect is a short-horizon negative coefficient (L3 ≈ −28 bp per 1 SD), consistent with pessimistic news reflecting transient bad-news pressure that is hard to detect amid daily noise. A likely reason the signal is weaker here is measurement/coverage differences—DNSI is a broad, dictionary-based index aggregated across sources, whereas Tetlock’s original tone measure is WSJ-specific and contextually financial, so added heterogeneity, timing misalignment, and measurement error can attenuate coefficients toward zero even if the underlying mechanism exists.


## 8. H2: Volume responds to extreme pessimism
Tetlock’s key volume prediction is that absolute sentiment forecasts higher volume. I implement the analogue using `DIA` detrended log volume (`vlm`) as a liquid Dow-linked proxy.


```python
# Volume regression
paper_v = paper.dropna(subset=["vlm"]).copy()
paper_v["abs_pess"] = paper_v["pessimism"].abs()

reg_v = prepare_regression_frame(paper_v, y_col="vlm", x_cols=["abs_pess", "ret_^DJI", "vlm"], lags=5).dropna()
yV = reg_v["vlm"]
XV = reg_v.drop(columns=["vlm"])
resV, resV_hac = ols_hac(yV, XV, hac_lags=5)

print(resV_hac.summary())
abs_lags = [f"abs_pess_L{k}" for k in range(1,6)]
print("\nJoint test (abs pessimism lags):", wald_joint_test(resV_hac, abs_lags))
```

### Interpretation (H2)
For H2, the volume regression shows that trading activity is highly predictable from its own dynamics and calendar effects ($R^2 \approx 0.815$), with strong persistence in vlm (significant positive lags 1–5) and pronounced day-of-week and January seasonality. Consistent with Tetlock’s intuition that sentiment primarily moves markets through attention/participation channels, the absolute pessimism terms have the expected interpretation (news intensity should raise volume), but in this DNSI-based implementation they are not jointly significant (joint p = 0.274) and only show weak, suggestive evidence at longer lags (e.g., L4 positive, p = 0.079). This pattern is plausible if DNSI “pessimism” blends many non-market news topics and introduces measurement error, which would attenuate the incremental effect of sentiment on volume once strong persistence and calendar regularities are controlled for.


## 9. H3: Small stocks (SMB / IWM–DIA spread)
I test whether pessimism has a more persistent negative effect on small stocks using:
1) the daily Fama–French **SMB** factor if available, else  
2) the return spread `ret_IWM − ret_DIA` as a transparent small-minus-large proxy.


```python
# Load Fama–French daily factors (SMB)
smb = None
try:
    ff = pdr.DataReader("F-F_Research_Data_Factors_daily", "famafrench", start="1984-01-01")[0] / 100.0
    ff.index = pd.to_datetime(ff.index)
    smb = ff["SMB"].rename("SMB")
except Exception as e:
    smb = None
    print("Could not load Fama–French daily factors:", e)

if smb is not None:
    panel = panel.join(smb, how="left")
    paper = panel.loc[(panel.index >= PAPER_START) & (panel.index <= PAPER_END)].copy()

# Construct proxy spread if needed
panel["ret_spread_IWM_DIA"] = panel.get("ret_IWM", np.nan) - panel.get("ret_DIA", np.nan)
paper = panel.loc[(panel.index >= PAPER_START) & (panel.index <= PAPER_END)].copy()
```

```python
target = "SMB" if ("SMB" in paper.columns and paper["SMB"].notna().sum() > 1000) else "ret_spread_IWM_DIA"
print("Small-stock target:", target)

reg_s = prepare_regression_frame(paper, y_col=target, x_cols=["pessimism", "ret_^DJI", "vlm", target], lags=5).dropna()
yS = reg_s[target]
XS = reg_s.drop(columns=[target])
resS, resS_hac = ols_hac(yS, XS, hac_lags=5)

print(resS_hac.summary())
pess_lags = [f"pessimism_L{k}" for k in range(1,6)]
print("\nJoint test (pessimism lags):", wald_joint_test(resS_hac, pess_lags))
print("Sum L1-L5:", coeff_sum_test(resS_hac, pess_lags))
```

### Interpretation (H3)
For H3, we test whether pessimistic news differentially impacts small stocks by regressing the SMB factor on lagged pessimism (plus lagged market returns, SMB lags, volume controls, and calendar effects). The pessimism coefficients are not individually significant and are jointly insignificant (joint p = 0.666; sum-of-lags p = 0.545), so we do not find strong evidence in this sample/specification that DNSI pessimism predicts subsequent small-minus-big returns. At the same time, the regression behaves sensibly: SMB is strongly related to lagged market moves (several ret_^DJI lags significant) and shows meaningful autocorrelation (SMB_L1 significant), indicating the model is capturing real factor dynamics rather than failing mechanically. A plausible interpretation is again due to broadness of measurement: an overall economic sentiment indicator (DNSI) will contain much less fine-grained stock information than a WSJ column that discusses specific stocks.


## 10. Nonlinearity diagnostic (LOWESS)
Tetlock reports that the return effect is stronger at extreme pessimism. I replicate the qualitative diagnostic by:
1) residualizing returns on controls **excluding** pessimism, then  
2) fitting LOWESS of residual returns on pessimism.


```python
def residualize_without_pessimism(df: pd.DataFrame, y_col: str, controls: list, lags: int = 5):
    reg = prepare_regression_frame(df, y_col=y_col, x_cols=controls, lags=lags)
    drop_cols = [c for c in reg.columns if c.startswith("pessimism_")]
    reg = reg.drop(columns=drop_cols, errors="ignore").dropna()
    y = reg[y_col]
    X = reg.drop(columns=[y_col])
    res, res_hac = ols_hac(y, X, hac_lags=lags)
    return reg.index, res.resid

idx, resid = residualize_without_pessimism(paper, "ret_^DJI", controls=["ret_^DJI", "vlm", "vol_proxy"], lags=5)
tmp = pd.DataFrame({"resid": resid}, index=idx).join(paper[["pessimism"]], how="left").dropna()

lw = lowess(tmp["resid"], tmp["pessimism"], frac=0.5, return_sorted=True)

import matplotlib.pyplot as plt
plt.figure(figsize=(7,4))
plt.scatter(tmp["pessimism"], tmp["resid"]*10000, alpha=0.2)
plt.plot(lw[:,0], lw[:,1]*10000, linewidth=2)
plt.xlabel("Pessimism (standardized)")
plt.ylabel("Residual Dow return (basis points)")
plt.title("LOWESS: Residual returns vs pessimism (paper-period)")
plt.show()
```

### Interpretation (Nonlinearity diagnostic)

The LOWESS diagnostic suggests little evidence of a strong nonlinear relationship between pessimism and next-day returns after removing the linear controls. It's noteworthy that the observation cloud is densest and effectively “centered” around standardized pessimism of roughly −1. This indicates the DNSI series (standardized in this pipeline) is shifted relative to a zero-mean baseline, so most trading days correspond to comparatively low measured pessimism and genuinely high-pessimism days are rarer. Even with that shift, residual returns remain centered near zero across the range and the LOWESS curve is nearly flat. Overall, the plot supports the view that the weak linear H1 findings are unlikely to be explained by missing curvature/threshold effects; instead they are more consistent with noise from the broader DNSI construction.

## 12. Overfitting and out-of-sample stability checks
To assess the probability that the original findings (or this replication) are overfit, I run:
1) **Time split:** estimate on an early subsample and evaluate coefficients/predictive signs on a later subsample.  
2) **Rolling estimation:** examine whether the sign and magnitude of the key next-day pessimism coefficient is stable over time.


```python
def fit_return_model(df: pd.DataFrame, start=None, end=None, lags=5):
    d = df.copy()
    if start is not None:
        d = d.loc[d.index >= pd.to_datetime(start)]
    if end is not None:
        d = d.loc[d.index <= pd.to_datetime(end)]

    # Try a sequence of specs to maximize usable observations
    specs = [
        ["pessimism","ret_^DJI","vlm","vol_proxy"],
        ["pessimism","ret_^DJI","vol_proxy"],
        ["pessimism","ret_^DJI","vlm"],
        ["pessimism","ret_^DJI"],
    ]

    last_err = None
    for xcols in specs:
        try:
            reg = prepare_regression_frame(d, y_col="ret_^DJI", x_cols=xcols, lags=lags)
            needed = ["ret_^DJI"] + [c for c in reg.columns if any(c.startswith(f"{x}_") for x in xcols)] + CAL_CONTROLS
            reg = reg.dropna(subset=needed)

            if reg.shape[0] <= (lags + 25):
                continue

            y = reg["ret_^DJI"]
            X = reg.drop(columns=["ret_^DJI"])
            _, res_hac = ols_hac(y, X, hac_lags=lags)
            return res_hac
        except Exception as e:
            last_err = e
            continue

    raise ValueError(f"No usable observations in [{start}, {end}] for any specification.") from last_err


# Split date: use 1992-01-01 if it lies inside the sample; otherwise fall back to sample midpoint
usable_idx = paper.dropna(subset=["pessimism","ret_^DJI"]).index
if len(usable_idx) == 0:
    raise ValueError("paper has no overlapping non-missing pessimism and ^DJI returns.")
sample_start = usable_idx.min()
sample_end = usable_idx.max()

desired_split = pd.Timestamp("1992-01-01")
if (desired_split <= sample_start) or (desired_split >= sample_end):
    split_ts = sample_start + (sample_end - sample_start) / 2
else:
    split_ts = desired_split

early_end = (split_ts - pd.offsets.BDay(1)).normalize()
m1 = fit_return_model(paper, start=sample_start, end=early_end)
m2 = fit_return_model(paper, start=split_ts, end=sample_end)
m1_params = pd.Series(m1.params, index=m1.model.exog_names)
m2_params = pd.Series(m2.params, index=m2.model.exog_names)

print("Sample start:", sample_start.date(), "Sample end:", sample_end.date())
print("Split date:", pd.Timestamp(split_ts).date(), "| Early end:", early_end.date())
print("Early period L1 coeff (bp):", basis_points(m1_params.get("pessimism_L1", np.nan)))
print("Late  period L1 coeff (bp):", basis_points(m2_params.get("pessimism_L1", np.nan)))
```

```python
# Rolling window stability of the key L1 coefficient
def rolling_L1(df: pd.DataFrame, window_years=5, step_months=6):
    d = df.dropna(subset=["pessimism","ret_^DJI"]).copy()
    start = d.index.min()
    end = d.index.max()
    out = []
    cur = start
    while cur + pd.DateOffset(years=window_years) <= end:
        w_end = cur + pd.DateOffset(years=window_years)
        try:
            m = fit_return_model(d, start=cur, end=w_end)
            params = pd.Series(m.params, index=m.model.exog_names)
            pvals = pd.Series(m.pvalues, index=m.model.exog_names)
            out.append((w_end, params.get("pessimism_L1", np.nan), pvals.get("pessimism_L1", np.nan)))
        except Exception:
            out.append((w_end, np.nan, np.nan))
        cur = cur + pd.DateOffset(months=step_months)
    return pd.DataFrame(out, columns=["window_end","coef_L1","pval_L1"]).set_index("window_end")

roll = rolling_L1(paper, window_years=5, step_months=6)
display(roll.tail(10))

import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
plt.plot(roll.index, 10000*roll["coef_L1"])
plt.axhline(0)
plt.title("Rolling 5-year estimate of pessimism_L1 effect on ^DJI returns")
plt.ylabel("Coefficient (basis points)")
plt.xlabel("Window end date")
plt.show()
```

### Interpretation (Overfitting and stability)

The stability checks suggest that short-horizon pessimism effects are not fully time-invariant in this sample and therefore at risk of being sample-specific. In the single split test, the estimated L1 pessimism coefficient falls sharply from about +20.8 bp per 1 SD in the early subsample to roughly +0.5 bp in the later subsample, indicating that the early relationship does not persist with the same magnitude out of sample. The rolling 5-year estimates reinforce this pattern: the L1 coefficient is positive throughout but monotonically declines from ~37–43 bp (1997 window ends) toward ~16 bp by 1999, while statistical significance weakens (p ≈ 0.045–0.086 early windows, then p > 0.30 thereafter). Overall, these results point to limited out-of-sample robustness for the return effect in this DNSI-based implementation, which is consistent with an economically small signal that is sensitive to the sample window and potentially diluted by measurement differences. There is still some evidence of a positive association in the earlier part of the period.

These diagnostics provide some evidence that the return-predictability channel can be sample-sensitive, which raises a moderate concern about overfitting in the original return specification. However, because this replication uses DNSI rather than WSJ-specific tone, a lot of the instability likely reflects measurement differences/attenuation rather than pure model overfitting. In code, I assess overfitting risk by re-estimating the same HAC(5) regression under a fixed pre-specified specification across (i) a time split and (ii) rolling 5-year windows, and comparing coefficient signs, magnitudes (bp per 1 SD), and p-values across subsamples.

### 13.2 Different asset classes
I test whether DNSI-based pessimism forecasts short-horizon returns (H1 analogue) for several liquid assets. For each asset, I run the same 5-lag regression:

$r^{asset}_t = \alpha + \sum_{k=1}^5 \gamma_k \text{pessimism}_{t-k} + \sum_{k=1}^5 \beta_k r^{asset}_{t-k} + \text{calendar controls} + \varepsilon_t,$

with HAC(5) inference. This isolates whether the return-pressure channel is equity-specific.


```python
ASSETS = {
    "S&P 500": "ret_^GSPC",
    "Nasdaq-100 (QQQ)": "ret_QQQ",
    "Long Treasuries (TLT)": "ret_TLT",
    "Gold (GLD)": "ret_GLD",
    "Bitcoin (BTC)": "ret_BTC-USD",
}

def asset_regression(df: pd.DataFrame, ret_col: str, lags=5):
    d = df.dropna(subset=[ret_col, "pessimism"]).copy()
    reg = prepare_regression_frame(d, y_col=ret_col, x_cols=["pessimism", ret_col], lags=lags).dropna()
    res, res_hac = ols_hac(reg[ret_col], reg.drop(columns=[ret_col]), hac_lags=lags)
    params = pd.Series(res_hac.params, index=res_hac.model.exog_names)
    pvals = pd.Series(res_hac.pvalues, index=res_hac.model.exog_names)
    pess_lags = [f"pessimism_L{k}" for k in range(1,lags+1)]
    out = {
        "coef_L1_bp": basis_points(params.get("pessimism_L1", np.nan)),
        "p_L1": pvals.get("pessimism_L1", np.nan),
        "joint_p": float(wald_joint_test(res_hac, pess_lags).pvalue),
        "sum_L1_L5_bp": basis_points(params.loc[pess_lags].sum()),
        "sum_p": float(coeff_sum_test(res_hac, pess_lags).pvalue),
        "N": int(res_hac.nobs),
    }
    return out

ext = panel.loc[panel.index >= "2000-01-01"].copy()
rows = []
for name, col in ASSETS.items():
    if col in ext.columns:
        rows.append((name, asset_regression(ext, col, lags=5)))
results_assets = pd.DataFrame({k:v for k,v in rows}).T
results_assets
```

### Interpretation (Asset-class extension)

Across asset classes, the DNSI-based return-pressure channel remains weak and largely indistinguishable from zero. For equities (S&P 500 and QQQ), the L1 effect is small and imprecise (≈ +8 to +11 bp per 1 SD pessimism, p ≈ 0.65), and the joint tests do not reject that the five pessimism lags are jointly zero (joint p ≈ 0.46–0.94). The most suggestive pattern appears for QQQ, where the sum of lags 1–5 is marginal (sum p ≈ 0.051) despite an economically small cumulative magnitude (~+3.7 bp), hinting that any effect is likely diffuse across days rather than concentrated at t+1. Non-equity assets show similarly weak evidence: TLT and BTC have negative L1 point estimates (−7.6 bp and −26.1 bp, respectively) but with large uncertainty and non-rejection in joint/sum tests, while GLD shows a small positive L1 with no significance. Overall, we see that DNSI pessimism does not robustly forecast short-horizon returns in a way that generalizes; if anything, the (limited) signal is most detectable in growth-heavy equities (where sentiment-sensitive risk-taking is highest) but still heavily diluted by noise.


## 14. Discussion of results and comparison to Tetlock

Overall, this replication using a DNSI-based pessimism measure finds much weaker return predictability than Tetlock (2007), but more supportive evidence for the attention/participation side.

In the paper-period return regression (H1), none of the five pessimism lags is individually significant and the joint and sum-of-lags tests are far from rejection, with economically modest, noisy coefficients; a LOWESS diagnostic similarly suggests no hidden nonlinearity that would rescue the signal. These results are consistent with either (i) the true return effect being small relative to daily noise and absorbed by standard controls, or (ii) dilution from measurement differences: DNSI is a broad, dictionary-constructed index aggregated across sources, whereas Tetlock’s tone measure is WSJ-specific and contextually financial.

In contrast, the volume tests (H2) show a clearer mechanism: while absolute pessimism is not strongly significant in the paper-period window, it becomes jointly significant post-2000, indicating that pessimistic news is more reliably associated with increased trading activity than with predictable next-day returns. This is closer in spirit to Tetlock’s interpretation that media tone moves markets primarily through investor attention and sentiment.

The small-stock test (H3) using SMB provides no statistically strong support for specific small-cap sensitivity, though the factor dynamics and market-return controls behave as expected. Stability checks show that any positive L1 return association declines over time and loses significance in later windows, suggesting limited out-of-sample robustness and reinforcing that the return channel is fragile under this proxy. Taken together, the split/rolling-window results suggest a moderate overfitting risk for return predictability in the original setting, though the weaker, less stable signal here is also consistent with attenuation from using a broader DNSI proxy rather than WSJ-specific tone.

Finally, extending H1 across assets (S&P 500, QQQ, TLT, GLD, BTC) yields no significant cross-market return predictability, with only a marginal cumulative effect for QQQ, consistent with the idea that any remaining signal is concentrated where sentiment-sensitive risk-taking is highest but still heavily diluted.

Overall, this replication study partially supports Tetlock’s broader mechanism that news tone correlates with market activity, but indicates that recovering his level of return predictability can be sensitive to using a WSJ-like, finance-focused tone measure and careful alignment, and is much harder to detect with a broader DNSI (economic sentiment) proxy.
